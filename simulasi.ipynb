```python
import numpy as np
import matplotlib.pyplot as plt

def dirac(n): 
    # dirac function
    return 1 if n==0 else 0

def dirac_vector(N):
    out=np.zeros(N)
    out[0]=1
    return out

def op1(signal):
    transformed_signal=np.zeros(np.size(signal))
    for t in np.arange(np.size(signal)):
        transformed_signal[t]=signal[t]-signal[t-1]
    return transformed_signal

def op2(signal):
    transformed_signal=np.zeros(np.size(signal))
    for t in np.arange(np.size(signal)):
        transformed_signal[t]=0.5*signal[t]+0.5*signal[t-1]
    return transformed_signal 

def op3(signal):
    transformed_signal=np.zeros(np.size(signal))
    for t in np.arange(np.size(signal)):
        transformed_signal[t]=0.7*transformed_signal[t-1]+signal[t]
    return transformed_signal

def op4(signal):
    transformed_signal=np.zeros(np.size(signal))
    for t in np.arange(np.size(signal)):
        transformed_signal[t]=1*transformed_signal[t-1]+signal[t]
    return transformed_signal

d=dirac_vector(20)
fig,ax=plt.subplots(2,2,sharex=True)
ax[0][0].stem(op1(d),label="Filter1")
ax[0][0].legend()
ax[0][1].stem(op2(d),label="Filter2")
ax[0][1].legend()
ax[1][0].stem(op3(d),label="Filter3")
ax[1][0].legend()
ax[1][1].stem(op4(d),label="Filter4")
ax[1][1].legend()
plt.suptitle("Impulse responses")
plt.show()

h=op4(op1(dirac_vector(20)))
plt.stem(h,label="Filter4(1)")
_=plt.axis([-5,20,0,1.2])
plt.show()
```

```python
# PERCOBAAN KEDUA
import numpy as np
import matplotlib.pyplot as plt

#untuk memberikan nilai terhadap suatu matrix yang ingin dibuat
def dirac(n): 
    # dirac function
    return 1 if n==0 else 0

#untuk membuat sebuah matrix zeros dengan ukuran N, dimana pada baris pertama diubah nilainya menjadi 1
def dirac_vector(N):
    out=np.zeros(N)
    out[0]=1
    return out

def op1(signal):
    print("ini size signal = " + str(np.size(signal)))
    transformed_signal=np.zeros(np.size(signal))
    print("ini transformed signal = " + str(transformed_signal))
    for t in np.arange(np.size(signal)):
        print("ini signal[t] = " + str(signal[t]))
        print("ini signal[t - 1] = " + str(signal[t-1]))
        print("ini hasil perhitungan = " + str(signal[t]-signal[t-1]))
        transformed_signal[t]=signal[t]-signal[t-1]
        print("transformed signal = " + str(transformed_signal))
    return transformed_signal

def op2(signal):
    print("ini size signal = " + str(np.size(signal)))
    transformed_signal=np.zeros(np.size(signal))
    print("ini transformed signal = " + str(transformed_signal))
    for t in np.arange(np.size(signal)):
        print("ini signal[t] = " + str(0.5*signal[t]))
        print("ini signal[t - 1] = " + str(0.5*signal[t-1]))
        print("ini hasil perhitungan = " + str(0.5*signal[t]+0.5*signal[t-1]))
        transformed_signal[t]=0.5*signal[t]+0.5*signal[t-1]
        print("transformed signal = " + str(transformed_signal))
    return transformed_signal 

def op3(signal):
    print("ini size signal = " + str(np.size(signal)))
    transformed_signal=np.zeros(np.size(signal))
    print("ini transformed signal = " + str(transformed_signal))
    for t in np.arange(np.size(signal)):
        print("ini signal[t] = " + str(signal[t]))
        print("ini transformed_signal[t - 1] = " + str(0.7*transformed_signal[t-1]))
        print("ini hasil perhitungan = " + str(0.7*transformed_signal[t-1]+signal[t]))
        transformed_signal[t]=0.7*transformed_signal[t-1]+signal[t]
        print("transformed signal = " + str(transformed_signal))
    return transformed_signal

def op4(signal):
    print("ini size signal = " + str(np.size(signal)))
    transformed_signal=np.zeros(np.size(signal))
    print("ini transformed signal = " + str(transformed_signal))
    for t in np.arange(np.size(signal)):
        print("ini signal[t] = " + str(signal[t]))
        print("ini transformed_signal[t - 1] = " + str(transformed_signal[t-1]))
        print("ini hasil perhitungan = " + str(1*transformed_signal[t-1]+signal[t]))
        transformed_signal[t]=1*transformed_signal[t-1]+signal[t]
        print("transformed signal = " + str(transformed_signal))
    return transformed_signal

d=dirac_vector(20)
fig,ax=plt.subplots(2,2,sharex=True)
print("==================OP1==================")    
ax[0][0].stem(op1(d),label="Filter1")
print("==================OP1==================\n")    
ax[0][0].legend()
print("==================OP2==================")    
ax[0][1].stem(op2(d),label="Filter2")
print("==================OP2==================\n")    
ax[0][1].legend()
print("==================OP3==================")    
ax[1][0].stem(op3(d),label="Filter3")
print("==================OP3==================\n")    
ax[1][0].legend()
print("==================OP4==================")    
ax[1][1].stem(op4(d),label="Filter4")
print("==================OP4==================\n")    
ax[1][1].legend()
plt.suptitle("Impulse responses")
plt.show()

h=op4(op1(dirac_vector(20)))
plt.stem(h,label="Filter4(1)")
plt.legend()
_=plt.axis([-5,20,0,1.2])
plt.show()
```

Proses yang terjadi pada kode instruksi "h=op4(op1(dirac_vector(20)))" secara garis besar adalah fungsi op4 atau filter 4 yang diberi input hasil dari nilai dirac_vector sebesar 20 yang telah masuk ke fungsi op1 atau filter 1 terlebih dahulu. Dirac_vector tersebut merupakan suatu matrix zero dengan jumlah elemen matrix sesuai dengan input yang diberikan, dimana pada dirac_vector ini memiliki nilai 1 pada elemen pertamanya saja sehingga dirac_vector(20) tersebut berarti terdapat matrix zero dengan jumlah 20 elemen matrix, dimana elemen pertama dari matrix tersebut adalah 1. Kemudian dirac_vector(20) ini dimasukkan pada filter 1 atau fungsi op1, dimana nantinya akan dilakukan perulangan sebanyak besar atau jumlah dari elemen pada dirac_vector(20) ini untuk melakukan pengurangan terhadap variabel signal (variabel yang digunakan untuk menyatakan input pada fungsi ini, dimana pada persoalan ini adalah dirac_vector(20)) pada index t dikurangkan dengan index t - 1 atau dapat dikatakan dikurangkan dengan melemen pada index sebelumnya. Hasil dari perhitungan ini kemudian dimasukkan ke dalam variabel transformed_signal sesuai dengan index dari perulangan atau index t, dimana variabel ini digunakan untuk menyimpan hasil dari signal atau dirac_vector(20) yang telah diberi filter. Kemudian hasil tersebut dimasukkan ke dalam fungsi op4 atau filter 4. Fungsi ini memiliki input dan sistematika perulangan yang sama dengan fungsi op1 atau filter 1, tetapi memiliki perhitungan yang sedikit berbeda, yaitu dengan melakukan pertambahan terhadap transformed_signal pada index t-1 dengan nilai signal pada index t dan kemudian nilai ini disimpan pada variabel transformed_signal pada index t.
